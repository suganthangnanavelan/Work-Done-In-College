```python
# Function to read the CSV file and return the text content
def read_csv_file(file_path):
    essay_text = ""
    with open(file_path, 'r') as file:
        for line in file:
            # Remove newline characters and quotes, then append to essay_text
            essay_text += line.strip().replace('"', '') + " "
    return essay_text

# Function to tokenize the text into words
def tokenize(text):
    words = text.split()
    return [word.strip(".,!?\"'()[]{}:;") for word in words]

# Function to perform frequency analysis
def frequency_analysis(words):
    frequency = {}
    for word in words:
        if word.lower() in frequency:
            frequency[word.lower()] += 1
        else:
            frequency[word.lower()] = 1
    return frequency

# Function to calculate word probabilities
def calculate_probabilities(frequency):
    total_words = sum(frequency.values())
    probabilities = {word: freq / total_words for word, freq in frequency.items()}
    return probabilities

# Load the essay text from the CSV file
essay_text = read_csv_file('E:/126156048/csv1.csv')

# Tokenize the text into words
words = tokenize(essay_text)

# Perform frequency analysis
word_freq = frequency_analysis(words)

# Calculate word probabilities
word_prob = calculate_probabilities(word_freq)

# Find the number of unique words
num_unique_words = len(word_freq)

# Display the number of unique words
print(f"Number of unique words: {num_unique_words}")

# Display the frequency of each word
print("Word Frequency Analysis:")
for word, freq in word_freq.items():
    print(f"{word}: {freq}")

# Display the probability of each word
print("Word Probability Analysis:")
for word, prob in word_prob.items():
    print(f"{word}: {prob:.4f}")

# Prepare data for plotting
import matplotlib.pyplot as plt

# Sort words by probability and get the top 10
sorted_word_prob = sorted(word_prob.items(), key=lambda x: x[1], reverse=True)
top_words = sorted_word_prob[:10]
words, probabilities = zip(*top_words)

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(words, probabilities, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Probability')
plt.title('Top 10 Words by Probability')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Define a list of common stop words
stop_words = set([
    'the', 'is', 'in' ,'In', 'and', 'to', 'of', 'for', 'on', 'with', 'a', 'an', 'it', 'that', 'as', 'by', 'at', 'or', 'from', 'was', 'which', 'are', 'be', 'will', 'has', 'have', 'had', 'not', 'you', 'this', 'but', 'we', 'they', 'can', 'if', 'has', 'more', 'other', 'than', 'so', 'up', 'out', 'over', 'under', 'into', 'its', 'been', 'are', 'all', 'some', 'would', 'also', 'such', 'their', 'our', 'about', 'these', 'those', 'where', 'when', 'why', 'how', 'each', 'both', 'few', 'many', 'most', 'much', 'only', 'most', 'then', 'here', 'there'
])
```

```python
# Function to perform frequency analysis
def frequency_analysis(words, stop_words):
    frequency = {}
    for word in words:
        if word not in stop_words and word != '':
            if word in frequency:
                frequency[word] += 1
            else:
                frequency[word] = 1
    return frequency
```

```python
# Tokenize the text into words
words = tokenize(essay_text)

# Perform frequency analysis excluding stop words
word_freq = frequency_analysis(words, stop_words)

# Find the number of unique words
num_unique_words = len(word_freq)
```

```python
# Display the number of unique words
print(f"Number of unique words: {num_unique_words}")
```

```python
# Prepare data for plotting
# Sort words by frequency and get the top 10
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
top_words = sorted_word_freq[:10]
words, frequencies = zip(*top_words)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Words by Frequency (Excluding Stop Words)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
